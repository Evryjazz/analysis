> **Bon voyons voir si on arrive à calculer celle ou celui qui a passé le plus grand nombre de coups de téléphone ??**
<br>*Bon pour se faire j'utilise une méthode 'un peu' doigt-levé-doigt-mouillé puisque je vais chercher les mots clés 'call|Call|appel|tel|Tel|téléphone|tél' dans les comments :*


```python
phone_number_perf = df["COMMENT"].str.contains("call|Call|appel|Appel|tel|Tel|téléphone|tél").groupby(df['WHO']).count()
phone_number_perf.order(ascending=False)
```

>**Whaaaat !?!?** Sachant que j'ai reporté 4000+ lignes, ça me semble peu probable... 
<br>
>Revoyons un peu ce calcul...


```python
df['is_call'] = df.COMMENT.str.contains("call|Call|appel|Appel|tel|Tel|téléphone|tél|telephonique")
is_call = df[df['is_call'] == True]
is_call_perf = is_call.groupby(is_call['WHO']).count()
final_is_call_perf = is_call_perf.loc[:,['is_call']].sort(columns='is_call', ascending=False)
final_is_call_perf
```

>Bon, ça ressemble plus à quelque chose... <br> 
>Donc a priori, **14%** (586/4068) de mes tâches sont liées à des coups de téléphone ! <br>
>**@PapaMalick** > persévère je pense que tu peux prendre la tête du podium !

>**Est-ce qu'on peut en déduire le temps que l'on a passé au tel ?**

```python
is_call_timing = is_call.groupby(is_call['WHO']).sum()
final_is_call_timing = is_call_timing.loc[:,['DURATION']].sort(columns='DURATION', ascending=False)
final_is_call_timing
```

**Climbing Information - #AmortLesAscenceurs**
<br>
>Bon en prenant en compte :
- le fait que je prends rarement l'ascenceur, qu'une marche mesure en [moyenne entre 17 et 21 cm](https://www.over-blog.com/Les_normes_pour_les_marches_descalier_conseils_et_recommandations-1095204284-art392075.html) (donc disons 19cm pour couper la poire en deux), 
- qu'il y a 101 marches pour atteindre nos locaux, 
- que je les montes deux fois par jour (oui faut bien aller manger),
- et que comme on l'a vu il y a 251 jours ouvrés entre février 2014 et le 31 janvier 2016. 

> **Combien j'ai grimpé de mètres depuis notre arrivée Quai de la Bastille ?**

```python
one_climb = 19*101
one_day_cm = 2*one_climb
one_day_m = one_day_cm/100
wasa_working_day = 251
overall_climbing = wasa_working_day*one_day_m

overall_climbing
```
